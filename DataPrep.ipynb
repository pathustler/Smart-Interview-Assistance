```python
import json


file_path = 'transcripts.json'
with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
import pandas as pd

df = pd.DataFrame([
    {
        'id': idx + 1,
        'question': item.get('question', ''),
        'ideal_answer': item.get('ideal_answer', '')
    }
    for idx, item in enumerate(data)
])
```

```python
import google.generativeai as genai
from google.colab import userdata
genai.configure(api_key=userdata.get('API_KEY'))
model = genai.GenerativeModel("gemini-2.0-flash-lite")
```

```python
def prompt(question):
    prompt = f"""
    The following is an interview question. Generate a bad, unprofessional, or vague answer to it in around 170 words.

    Question: {question}

    Bad Answer:
    """
    response = model.generate_content(prompt)
    return response.text.strip()
```

```python
df["bad_answer"] = df["question"].apply(prompt)

```

```python
import pandas as pd


s1,s2,s3,s4 = pd.read_csv("Sample_Part_1.csv"),pd.read_csv("Sample_Part_2.csv"),pd.read_csv("Sample_Part_3.csv"),pd.read_csv("Sample_Part_4.csv")

dataset = pd.concat([s1,s2,s3,s4])
```

```python
dataset.set_index('Unnamed: 0', inplace=True)
```

```python
dataset
```

```python
ideal_answers = dataset.drop(columns=['bad_answer'])
bad_answers = dataset.drop(columns=['ideal_answer'])
```

```python
ideal_answers['professionalism_rating'] = 1.0
bad_answers['professionalism_rating'] = 0.0
```

```python
ideal_answers.rename(columns={'ideal_answer': 'answer'}, inplace=True)
bad_answers.rename(columns={'bad_answer': 'answer'}, inplace=True)
```

```python
final_dataset = pd.concat([ideal_answers, bad_answers])
```

```python
final_dataset = final_dataset.sample(frac=1).reset_index(drop=True)
```

```python
final_dataset = final_dataset.drop(columns=['id'])
```

```python
final_dataset.to_csv('RatedInterviewQuestionsDataset.csv')
```

```python
def rate_answer(row):
    question = row['question']
    answer = row['answer']

    prompt = f"""
    You are an expert interview coach evaluating the quality of answers in a professional interview setting.

    Given the following question and candidate's answer, rate the **professionalism** of the answer on a scale from **0 to 1**, where:
    - 1 means highly professional, well-structured, and perfectly relevant.
    - 0 means unprofessional, vague, or completely off-topic.

    **Consider the following criteria:**
    - Clarity and tone of communication
    - Relevance to the question
    - Depth and specificity of response
    - Appropriateness for a professional interview context

    **Important:** Only return a single decimal value between **0 and 1**. Do not include any explanation or extra text.

    Question: {question}

    Answer: {answer}
    """
    response = model.generate_content(prompt)
    return response.text.strip()
```

```python
import time
```

```python
final_rated_df = pd.DataFrame()
```

```python

for i in range(0,200,25):
  small_df = final_dataset.iloc[i:i+25,:]
  small_df["generated_rating"] = small_df.apply(prompt, axis=1)
  final_rated_df = pd.concat([final_rated_df, small_df], ignore_index=True)
  time.sleep(60)
```
